```python
from __future__ import print_function
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
matplotlib.rcParams.update({'font.size': 18})
```

```python
def load_data():
    return pd.read_csv("Advertising.csv")

def plot_original_data(X, y):
    plt.scatter(X, y, color='#1f77b4', marker='o')

    plt.xlabel("Spendings, M$")
    plt.ylabel("Sales, Units")
    plt.title("Sales as a function of radio ad spendings.")
    plt.show()
```

```python

def update_w_and_b(X, y, w, b, alpha):
    dl_dw = 0.0
    dl_db = 0.0
    N = len(X)

    dl_dw = -2 * np.sum(X * (y - (w*X + b)))
    dl_db = -2 * np.sum(y - (w*X + b))

    # update w and b
    w = w - (dl_dw/N) * alpha
    b = b - (dl_db/N) * alpha

    #print("w:", w, "b:", b)

    return w, b
```

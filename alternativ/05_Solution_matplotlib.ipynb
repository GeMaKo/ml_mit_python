```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# increase general font size of plot labels
plt.rcParams.update({'font.size': 14})
```

# Lade Beispieldatenset

```python
# Lade Datenset
taxis_df = sns.load_dataset('taxis')
```

**Workflow**
1. Prepare Data: Nehmen wir an, die Daten sind vorbereitet.
2. Create Plot 
3. Plot 
4. Customized Plot 
5. Save Plot 
6. Show Plot

*Weitere Cheat Sheets auf unterschiedlichen Niveaus: https://matplotlib.org/cheatsheets/*

```python
# 2. create plot
fig = plt.figure()

# 3. plot 
ax = fig.add_subplot(111)

# 4. customize
ax.scatter(taxis_df.total, taxis_df.distance, marker = 'x') #, color = 'green', linewidth = 3)
plt.xlabel('total ($)')
plt.ylabel('distance (miles)')
plt.title('My first Plot')
ax.set_xlim(-1, 200) # Passe Achsen an auf 30x100 Fenster
ax.set_ylim(-1, 50)

# 5. save
plt.savefig('test.png')

# 6.show
plt.show()
```

# Barplot

```python
# 2. create plot
fig = plt.figure()

# 3. plot 
ax = fig.add_subplot(111)

# 4. customize
ax.hist(taxis_df.color) #, color = 'green', linewidth = 3)
plt.xlabel('color')
plt.title('My first Bar Plot')
#ax.set_xlim(-1, 50) # Passe Achsen an auf 30x100 Fenster

# 5. save
plt.savefig('test.png')

# 6.show
plt.show()
```

# Aufgabe: Boxplot

```python
# 2. create plot
fig = plt.figure()

# 3. plot 
ax = fig.add_subplot(111)

# 4. customize
ax.boxplot([taxis_df[taxis_df['color'] == 'yellow']['total'],
            taxis_df[taxis_df['color'] == 'green']['total']]) #, marker = 'x', color = 'green')
plt.ylabel('total ($)')
plt.title('My first boxplot')
plt.xticks([1, 2], ['yellow', 'green'])

# 5. save
plt.savefig('test.png')

# 6.show
plt.show()
```

## Subplots

```python

data = {'apple': 10, 'orange': 15, 'lemon': 5, 'lime': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)
fig.suptitle('Categorical Plotting')
```

# Aufgabe: Linienplot

```python
# 1. Prepare Data
taxis_df['pickup_date'] = pd.to_datetime(taxis_df['pickup']).dt.date
pickups_by_day = taxis_df['pickup_date'].value_counts().sort_index()
#pickups_by_day

taxis_df['dropoff_date'] = pd.to_datetime(taxis_df['dropoff']).dt.date
dropoff_by_day = taxis_df['dropoff_date'].value_counts().sort_index()
#dropoff_by_day
```

```python
# 2. create plot
fig = plt.figure(figsize=(14,7))

# 3. plot 
ax = fig.add_subplot(111)

# 4. customize
ax.plot(pickups_by_day, label = 'Pickups') #, linewidth = 3, color = 'green')
ax.plot(dropoff_by_day, label = 'Dropoffs') #, linewidth = 3, color = 'green')
plt.xlabel('Date')
plt.ylabel('Pickups')
plt.title('My first Line Plot')
plt.xticks(rotation=15)
plt.legend(loc = 'best') #No handles with labels found to put in legend.

# 5. save
plt.savefig('test.png')

# 6.show
plt.show()
```

# DataFrame-Methoden: Spaltennamen ändern, Werte ersetzen, sortieren, gruppieren, Zufallszeile, Pivottabelle

## Spaltennamen ändern

```python
df_small.rename(columns = {'pickup':'abholzeit', 'dropoff':'absetzzeit'}, inplace = True)
df_small.head()
```

## Werte ersetze

```python
df_small.replace('yellow', 'gelb').head()
```

```python
df_small.loc[0,'distance'] = 2.3
df_small.head()
```

```python
#df_small.fillna(df3_small.mean()) 
```

## Sortieren

```python
df_neusortiert = df_small.sort_values(by = 'fare')
df_neusortiert.head()
```

```python
df_neusortiert.sort_index().head()
```

## Gruppieren

```python
df_grouped = df_small.groupby('payment').mean()
df_grouped
```

```python
df_gr = df_small.groupby('payment').agg({'distance': ['mean','max'], 'total': 'mean'})
df_gr
```

```python
sns.scatterplot(x =  df_gr['distance', 'mean'], y = df_gr['total', 'mean'])
```

## Pivottabelle und Melting

```python
df_grouped
```

### Melting

```python
df_grouped['payment'] = df_grouped.index
df_grouped
```

```python
df_new = pd.melt(df_grouped, id_vars=["payment"],
                 value_vars=["distance","fare","tip","tolls","total"],
                 value_name="values")
df_new
```

### Pvotisieren

```python
df_back = df_new.pivot(index=["payment"],
                 columns=["variable"],
                 values="values")
df_back
```

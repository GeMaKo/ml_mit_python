## Daten, die in Paketen geliefert werden


```python
# Lade Paket (z.B. seaborn)
import seaborn as sns
```

```python
# Lade Datenset
taxis_df = sns.load_dataset('taxis')
```

```python
taxis_df
```

## Daten speichern

### Als Csv

```python
taxis_df.to_csv('taxis.csv', sep=' ', index=False)
```

```python
help(taxis_df.to_csv)
```

### <span style='color:Green'> *Aufgabe*: Als Excel  <span> 

```python
!pip install openpyxl
```

```python
taxis_df.to_excel('taxis.xlsx', header=False)
help(taxis_df.to_csv)
```

### Mit Pickle

```python
import pickle
```

```python
filename = 'taxis_pickle'
```

```python
outfile = open(filename, 'wb')
```

```python
help(open)
```

```python
pickle.dump(taxis_df,outfile)
outfile.close()
```

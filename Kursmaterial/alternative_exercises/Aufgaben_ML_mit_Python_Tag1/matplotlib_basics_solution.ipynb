# Matplotlib Basics

### 1. Begin with importing pandas and seaborn again but also add matplotlib this time the following way: `matplotlib.pyplot as plt`.`

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crash_df = sns.load_dataset('car_crashes')
```

### Variable Explanations

| Variable | Explanation |
| --- | --- |
| total | Number of drivers involved in fatal collisions per billion miles |
| speeding | Number of drivers involved in fatal collisions per billion miles Who Were Speeding |
| alcohol | Number of drivers involved in fatal collisions per billion miles Who Were Alcohol-Impaired |
| not_distracted | Number of drivers involved in fatal collisions per billion miles Who Were Not Distracted |
| no_previous | Number of drivers involved in fatal collisions per billion miles Who Had Not Been Involved In Any Previous Accidents |
| ins_premium | Car Insurance Premiums (\$) |
| ins_losses | Losses incurred by insurance companies for collisions per insured driver (\$) | 
| abbrev | State |

https://www.kaggle.com/fivethirtyeight/fivethirtyeight-bad-drivers-dataset

Note that categories are not exclusive. A driver might have been drinking alcohol, while speeding, without having had previous car accidents when entering the statistic as fatal collision.

### Let us add the regions indicating which region a state belongs to to our dataframe and create a grouped dataframe that we are going to work with first.

```python
crash_df['US Region'] = ['South', 'West', 'West', 'South', 'West', 'West', 'Northeast', 'South', 'South', 'South', 'South', 'West', 'West', 'Midwest', 'Midwest', 'Midwest',
'Midwest', 'South', 'South', 'Northeast', 'South', 'Northeast', 'Midwest', 'Midwest', 'South', 'Midwest', 'West', 'Midwest', 'West', 'Northeast', 'Northeast', 'West', 'Northeast', 'South', 'Midwest', 'Midwest', 'South', 'West', 'Northeast', 'Northeast', 'South', 'Midwest',
'South', 'South', 'West', 'Northeast', 'South', 'West', 'South', 'Midwest', 'West']
```

```python
crash_df_regions = crash_df.groupby('US Region').mean().reset_index()
```

### 2.1 Let us start with a simple bar plot showing the 'total' values for each region of `crash_df_regions`. There are many ways one can do this, choose the one you prefer. One option is to use matplotlib's `plt.bar()` function. To do so, select the right labels for x and y and make sure to add your dataframe with the `data` keyword

```python
plt.bar('US Region', 'total', data=crash_df_regions)
plt.show()

# alternatives
# crash_df_regions.plot(kind='bar', x='US Region', y='total')
# # plt.tick_params(axis='x', colors='white')
# # plt.tick_params(axis='y', colors='white')

# state_abbreviations = crash_df_regions['US Region']
# total_collisions = crash_df_regions['total']
# plt.bar(state_abbreviations, total_collisions)
```

### 2.2 Let us add some more information to make it easier for someone else to understand what the data is representing. So first, add labels for x and y axes as well as a title using the matplotlib pyplot functions `xlabel()`,  `ylabel()`, and `title()`.

```python
plt.bar('US Region', 'total', data=crash_df_regions)
plt.xlabel('US Regions')
plt.ylabel('total collisions')
plt.title('Total collisions among US Regions per billion miles')

plt.show()

```

### 2.3 There are a lot of options to adjust your plot. You can find them all in the [documention](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html). For now adjust the `yticks` of the y axis so they are only in steps of 5, i.e., [5, 10, 15, 20]. Also add a `color` argument to the plt.bar function and add a list of 4 different colors.
### At the end, if you are happy with the result, you can also save the plot with `plt.savefig()` and a file name or path.

```python
plt.bar('US Region', 'total', data=crash_df_regions, color=['red', 'blue', 'green', 'yellow'])
plt.xlabel('US Regions')
plt.ylabel('total collisions')

plt.title('Total collisions among US Regions per billion miles')

plt.yticks([5, 10, 15, 20])

plt.savefig('total_us_regions_barplot.png')
plt.show()
```

### 4.1 Let's try to make a different kind of graph. The `plt.plot()` function creates a line plot that connects all given datapoints to a line. In the following we are going to use the original car crash dataset but first we sort it according to the insurance premium. Use `sort_values` and save the returned dataframe in a new variable. Afterwards create a line plot, using the sorted dataframe, in a similar way to the bar plot before, using 'abbrev' and 'ins_premium' and add x labels, y labels, and a title. 

### Use `plt.figure()` and change the `figsize` parameter so the plot is wide enough for state name abbreviation to be readable. <br>Tip: Make sure to set the figure before creating any plot.  

```python
crash_df_sorted_premium = crash_df.sort_values('ins_premium')

plt.figure(figsize=(20, 6))
plt.plot('abbrev', 'ins_premium', data=crash_df_sorted_premium)

plt.xlabel('US States')
plt.ylabel('Insureance premium ($)') 
plt.title('Insurance premiums')

plt.show()
```
